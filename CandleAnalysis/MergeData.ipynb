### Iterate through the data folder and take the first option expiration date of each

```python
import os
import pandas as pd
import numpy as np
```

```python
directory = r"C:\Users\Victor\Desktop\0-Leetcode\Desktop\Projects\CandleAnalysis\2021-data"
#savePath = 
dataList = []

for filename in os.listdir(directory):
    #print(filename)
    insideDir = directory + "\\" + filename + "\SPY"
    #print(insideDir)
    nextExpirationDate = os.listdir(insideDir)[0]
    #print(nextExpirationDate)
    df = pd.read_csv(os.path.join(insideDir, nextExpirationDate))
    #npData = df.to_numpy()
    #print(npData)
    
    #merge the dataset
    dataList.append(df)
dfAll = pd.concat(dataList,axis=0,ignore_index=True)
dfAll
```

```python
dfAll.to_csv(r"C:\Users\Victor\Desktop\0-Leetcode\Desktop\Projects\CandleAnalysis\mergedData.csv")
```
