```python
#fucntion for rounding up - https://kodify.net/python/math/round-decimals/
def round_decimals_up(number:float, decimals:int=2):
    """
    Returns a value rounded up to a specific number of decimal places.
    """
    if not isinstance(decimals, int):
        raise TypeError("decimal places must be an integer")
    elif decimals < 0:
        raise ValueError("decimal places has to be 0 or more")
    elif decimals == 0:
        return math.ceil(number)

    factor = 10 ** decimals
    return math.ceil(number * factor) / factor
round_decimals_up(15.223,2)
```

## We find the strike price for the call and the put and we sum those prices to get the strangle price

```python
#constant: time="9:45:00 AM"
#varaibles: date,callput,strikePrice,
#output: openTradePriceSum (call and put together)

#df3=dataframe with just time="9:45:00 AM"
time = "9:45:00 AM"
df3 = df.loc[df['TimeBarStart'] == time]

#takes in rowIndex & Date & Time, outputs Max Under Close Ask Price & lastIndex for date
def dateTimeToMaxUnderCloseAskPrice(startIndex:int,date:int,time:str) -> (int,int):
    rows = df3.loc[(df3['Date'] == date) & (df3['TimeBarStart'] == time)] #note- can be optimized by breaking when the date/time changes
    #print("rows: {}".format(rows))
    if rows['UnderCloseAskPrice'].empty:
        return (-1,startIndex)
    maxUnderCloseAskPrice = max(rows['UnderCloseAskPrice'])
    lastIndex = startIndex + len(rows) - 1
    
    #print(maxUnderCloseAskPrice)
    return (maxUnderCloseAskPrice,lastIndex)

printCount = 0
dateTimeMapToMUCAP = {} #map[(date,time)] = MUCAP
df3Len = len(df3)
index = 0
df3ResultList = [] #[index,date,time,put_SP,call_SP,callPrice,putPrice,sumCallPut]
while index < df3Len:
    row = df3.iloc[index]
    #print('row: {}'.format(row))
    startIndex = index
    
    date, strike = row['Date'], row['Strike']

    #get MUCAP price to round for call/put prices
    MUCAP,lastIndex = dateTimeToMaxUnderCloseAskPrice(index,date,time)
    dateTimeMapToMUCAP[(date,time)] = MUCAP

    put_SP = int(MUCAP)
    call_SP = math.ceil(MUCAP)

    #get openTradePrice for the put_SP we calculated
    callPut = "P"
    openTradePriceDF_put = df3.loc[ (df3['TimeBarStart'] == time) & (df3['CallPut'] == callPut) & (df3['Date']== date) & (df3['Strike']==put_SP)]
    openTradePrice_put = openTradePriceDF_put.iloc[0]['OpenTradePrice']
    #get openTradePrice for the call_SP we calculated
    callPut = "C"
    openTradePriceDF_call = df3.loc[ (df3['TimeBarStart'] == time) & (df3['CallPut'] == callPut) & (df3['Date']== date) & (df3['Strike']==call_SP)]
    openTradePrice_call = openTradePriceDF_call.iloc[0]['OpenTradePrice']
    
    
    sumCallPut = openTradePrice_put + openTradePrice_call
    df3ResultList.append([index,date,time,put_SP,call_SP,openTradePrice_call,openTradePrice_put,round_decimals_up(sumCallPut,2)])
    
    if printCount <10:
        print("index={},date={}, time={}, put_SP={},call_SP={}, callPrice={}, putPrice={}, sumCallPut={}".format(index,date,time,put_SP,call_SP,openTradePrice_call,openTradePrice_put,round_decimals_up(sumCallPut,2)))
        printCount += 1
    
    index = lastIndex + 1
```
