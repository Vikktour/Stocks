```python
column_values = df["Date"].values.ravel()
unique_values = pd.unique(column_values)
#print(unique_values)
```

# The max value of the stock price at a given date and 9:45am

```python
currentDate = df1['Date'][0]
currentTime = df1['TimeBarStart'][0]
rows = df1.loc[(df1['Date'] == currentDate) & (df1['TimeBarStart'] == currentTime)]
#print("rows: {}".format(rows))
maxUnderCloseAskPrice = max(rows['UnderCloseAskPrice'])
print(maxUnderCloseAskPrice)
```

# Convert Date Time to the Max Under Close Ask Price

```python
#takes in Date & Time, outputs Max Under Close Ask Price
def dateTimeToMaxUnderCloseAskPrice(date:int,time:str) -> int:
    rows = df1.loc[(df1['Date'] == date) & (df1['TimeBarStart'] == time)]
    #print("rows: {}".format(rows))
    if rows['UnderCloseAskPrice'].empty:
        return -1
    maxUnderCloseAskPrice = max(rows['UnderCloseAskPrice'])
    
    
    #print(maxUnderCloseAskPrice)
    return maxUnderCloseAskPrice
```
